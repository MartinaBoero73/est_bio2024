A partir del paper "Cryptic inoviruses revealed as pervasive in bacteria and archaea across Earth’s biomes", realizamos un análisis exploratorio de los inovirus hallados dentro del genoma de Campylobacter

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("./Datos/Tabla2.xlsx", header=1, sheet_name=2)
```

```python
df
```

```python
datos_campy = df[df["Dataset name"].str.contains("Campylobacter", na=False)]

datos_campy
```

```python
total_bp_inovirus = datos_campy['Total bp inoviruses']

plt.hist(total_bp_inovirus, bins=20, color='red', edgecolor='black')
plt.xlabel('Total bp inoviruses')
plt.ylabel('Frecuencia')
plt.title('Distribución de la longitud total de las secuencias de inovirus dentro de Campylobacter')
plt.show()
```
